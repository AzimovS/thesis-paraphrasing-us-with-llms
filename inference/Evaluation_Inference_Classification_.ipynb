```python
df = pd.read_csv('llama2-1.csv')
```

```python
df
```

```python
filter_col = [col for col in df if col.startswith('res')]
df = df[filter_col]
df
```

```python
df.mean()
```

```python
df.mean().mean()
```

```python
filter_col = [col for col in df if col.startswith('res increase')]
df_increase = df[filter_col]
df_increase.mean()
```

```python
df_increase.mean().mean()
```

```python
filter_col = [col for col in df if col.startswith('res decrease')]
df_decrease = df[filter_col]
df_decrease.mean()
```

```python
df_decrease.mean().mean()
```

```python
filter_col = [col for col in df if col.startswith("res don't change")]
df_dontchange = df[filter_col]
df_dontchange.mean()
```

```python
df_dontchange.mean().mean()
```

```python
filter_col = [col for col in df if col.startswith("res inc")] + [col for col in df if col.startswith("res dec")]
df_without_dontchange = df[filter_col]
df_without_dontchange.mean()
```

```python
df_without_dontchange.mean().mean()
```
